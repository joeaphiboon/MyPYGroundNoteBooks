กำหนดต่า Library Package ที่ต่้องใช้งาน

```python
from pandas_datareader import data as pdr
import pandas as pd
import fix_yahoo_finance as yf
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
%matplotlib inline
```

กำหนด Function ในการเรียกราคาหุ้นจาก Yahoo Finance

```python
def GetStock(stockname, start, end) :
    return (pdr.get_data_yahoo(stockname, start, end))
```

```python
symbols=['GC=F','EURUSD=X','JPY=X','THB=x']
```

เรียกใช้ Function ในการดึงข้อมูลกราฟตามช่วงเวลาที่กำหนดมาแสดง

```python
Dataf1=GetStock(symbols[0],'2015-3-7','2017-3-7')

Dataf2=GetStock(symbols[1],'2015-3-7','2017-3-7')

Dataf3=GetStock(symbols[2],'2015-3-7','2017-3-7')

Dataf4=GetStock(symbols[3],'2015-3-7','2017-3-7')
```

```python
plt.figure(1)
plt.subplot(111)
Dataf1['Close'].plot(figsize=(12,9))
plt.title(symbols[0])

plt.figure(2)
plt.subplot(111)
Dataf2['Close'].plot(figsize=(12,9))
plt.title(symbols[1])

plt.figure(3)
plt.subplot(111)
Dataf3['Close'].plot(figsize=(12,9))
plt.title(symbols[2])

plt.figure(4)
plt.subplot(111)
Dataf4['Close'].plot(figsize=(12,9))
plt.title(symbols[3])

plt.show
```

Plot ค่า Percentile ที่ 95% และ 5%

```python
plt.figure(5)
plt.figure(figsize=(8,5))
plt.subplot(111)
sns.distplot(Dataf1['Close'])
p=np.percentile(Dataf1['Close'],5)
q=np.percentile(Dataf1['Close'],95)
plt.axvline(x=p, linewidth=3,color='r')
plt.axvline(x=q, linewidth=3,color='g')
plt.figtext(0.2,0.6,"percentile (0.05): %.4f" % p,fontsize=14)
plt.figtext(0.53,0.5,"percentile (0.95): %.4f" % q,fontsize=14)
plt.title(symbols[0])

#############
plt.figure(6)
plt.figure(figsize=(8,5))
plt.subplot(111)
sns.distplot(Dataf2['Close'])
p=np.percentile(Dataf2['Close'],5)
q=np.percentile(Dataf2['Close'],95)
plt.axvline(x=p, linewidth=3,color='r')
plt.axvline(x=q, linewidth=3,color='g')
plt.figtext(0.2,0.6,"percentile (0.05): %.4f" % p,fontsize=14)
plt.figtext(0.53,0.5,"percentile (0.95): %.4f" % q,fontsize=14)
plt.title(symbols[1])
```
