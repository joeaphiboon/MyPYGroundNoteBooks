สร้าง function คำนวนหา %change ของราคาเทียบกับวันก่อนหน้า

```python
def priceChange (price):
    return (price-price.shift(1))/price.shift(1)
```

```python
setChange= priceChange (df1['SET50'])
BBLChange= priceChange (df1['BBL'])
KBANKChange= priceChange (df1['KBANK'])
SCBChange= priceChange (df1['SCB'])
KTBChange= priceChange (df1['KTB'])
```
