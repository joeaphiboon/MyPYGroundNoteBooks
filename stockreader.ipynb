การนำเข้าข้อมูลหุ้นจาก Yahoo Finance 

```python
from pandas_datareader import data as pdr
import fix_yahoo_finance as yf
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
stockname='ADVANC.BK'
data = pdr.get_data_yahoo(stockname, start="2017-01-01", end="2017-04-30")
data.head()
```

```python
fig, ax = plt.subplots()
ax.scatter(data.index, data['Close'], color = 'green', label = 'Close Price') # scatter plot
ax.plot(data['Close']) #plot line graph

plt.xlabel('Stock Price')
plt.ylabel('Price')
plt.grid(True)
plt.legend()
plt.show()
```

```python
fig, ax1 = plt.subplots()
ax1 = plt.subplot2grid((6,1),(0,0), rowspan =5, colspan=1)
ax2 = plt.subplot2grid((6,1),(5,0), rowspan =1, colspan=1, sharex = ax1)
ax1.plot(data.index, data['Close'])
ax2.plot(data.index, data['Volume'])
plt.show()
```
